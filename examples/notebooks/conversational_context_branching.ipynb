# Conversational Context Branching

Demonstrate how memoir enables safe exploration of hypothetical scenarios without corrupting the main memory timeline.

**Scenario**: User asks "What if I had chosen career path X instead of Y?"

```python
import asyncio
from memoir.core.memory import ProllyTreeMemoryStoreManager
from memoir.classifier.intelligent import IntelligentClassifier
from langchain_openai import ChatOpenAI
```

```python
# Initialize memory manager with intelligent classification
llm = ChatOpenAI(model="gpt-4", temperature=0)
classifier = IntelligentClassifier(llm=llm)
memory = ProllyTreeMemoryStoreManager(classifier=classifier)
```

```python
# Build main timeline: User's actual career path
await memory.store_memory(
    "I graduated with a computer science degree in 2020",
    metadata={"message": "Initial career foundation"}
)

await memory.store_memory(
    "I chose to work at a startup doing backend development",
    metadata={"message": "Career choice: startup backend"}
)

await memory.store_memory(
    "I've been promoted to senior engineer after 3 years",
    metadata={"message": "Career progression"}
)

main_commit = await memory.get_current_commit()
```

```python
# Create branch for hypothetical scenario
await memory.create_branch("alternative_path")
await memory.checkout("alternative_path")

print(f"Created branch 'alternative_path' from commit: {main_commit[:8]}")
```

```python
# Explore alternative: Big Tech career path
await memory.store_memory(
    "What if I had joined Google as a frontend engineer instead?",
    metadata={"message": "Alternative career exploration"}
)

await memory.store_memory(
    "At Google, I would have focused on React and large-scale UI systems",
    metadata={"message": "Alternative skillset development"}
)

await memory.store_memory(
    "I might have become a tech lead for YouTube's frontend team",
    metadata={"message": "Alternative career progression"}
)

alt_commit = await memory.get_current_commit()
```

```python
# Search in alternative timeline
alt_results = await memory.search_memories("my career skills")
print("Alternative timeline skills:")
for result in alt_results[:3]:
    print(f"- {result.content}")
```

```python
# Switch back to main timeline
await memory.checkout("main")

main_results = await memory.search_memories("my career skills")
print("\nMain timeline skills:")
for result in main_results[:3]:
    print(f"- {result.content}")
```

```python
# Compare timelines side by side
print(f"Main timeline commit: {main_commit[:8]}")
print(f"Alternative timeline commit: {alt_commit[:8]}")
print("\nKey insight: Safe exploration without memory corruption")
```

```python
# Cleanup: Delete experimental branch
await memory.delete_branch("alternative_path")
print("Alternative branch deleted, main timeline preserved")
```
